```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
data_variant = 7
```

```python
data = np.loadtxt(f'seismic_test_dataset{data_variant}.csv', delimiter=',', skiprows=1)
```

```python
print(data.shape, data)
```

```python
spectral_density = np.abs(np.fft.fft(data[:, 1]))**2
f_sampling = (data.shape[0] - 1) / (data[-1, 0] - data[0, 0])

print(f_sampling)
```

```python
fig = plt.figure()

ax = fig.add_subplot()
ax.plot(data[:, 0], data[:, 1])
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.linspace(0, f_sampling, len(spectral_density)), spectral_density)
```
